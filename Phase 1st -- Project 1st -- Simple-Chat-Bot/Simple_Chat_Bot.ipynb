```python
import random
import datetime

class AdvancedChatbot:
    def __init__(self):
        self.context = {}
        self.responses = {
            "who are you": "I'm a sophisticated chatbot designed to assist you with various tasks.",
            "Who are you": "I'm a sophisticated chatbot designed to assist you with various tasks.",
            "how are you": "I'm just a program, but I'm always here to help!",
            "How are you": "I'm just a program, but I'm always here to help!",
            "what's your name": "You can call me Chatbot!",
            "What's your name": "You can call me Chatbot!",
            "what's the time": f"It's currently {datetime.datetime.now().strftime('%H:%M')}.",
            "What's the time": f"It's currently {datetime.datetime.now().strftime('%H:%M')}.",
            "what's your favorite color": "I don't have a favorite color. I'm more into 0s and 1s.",
            "What's your favorite color": "I don't have a favorite color. I'm more into 0s and 1s.",
            "tell me a joke": self._tell_joke,
            "Tell me a joke": self._tell_joke,
            "what's the weather like": self._get_weather,
            "What's the weather like": self._get_weather,
            "play game": self._play_game,
            "Play game": self._play_game,
            "ask user question": self._ask_user_question,
            "Ask user question": self._ask_user_question,
        }
        self.normal_questions = [
            "How was your day?",
            "What are your plans for the weekend?",
            "Have you watched any good movies lately?",
        ]
        self.farewell = "Goodbye! Have a great day."

    def greet(self):
        return "Hello! I'm Chatbot. How can I assist you today?"

    def handle_input(self, user_input):
        user_input = user_input.lower()
        if user_input in self.responses:
            if callable(self.responses[user_input]):
                return self.responses[user_input]()
            else:
                return self.responses[user_input]
        elif user_input.startswith("ask user question"):
            return self._ask_user_question(user_input)
        else:
            return self._handle_default_response(user_input)

    def ask_question(self, question):
        print(question)
        answer = input()
        self.context[question] = answer

    def ask_questions(self):
        for question in self.context.keys():
            self.ask_question(question)

    def _handle_default_response(self, user_input):
        if 'thank' in user_input:
            return "You're welcome!"
        elif 'bye' in user_input:
            return "Goodbye! Take care."
        else:
            return "I'm sorry, I didn't quite understand that. Can you please rephrase?"

    def _tell_joke(self):
        jokes = [
            "Why don't scientists trust atoms? Because they make up everything!",
            "Parallel lines have so much in common. It’s a shame they’ll never meet.",
            "I told my wife she was drawing her eyebrows too high. She looked surprised.",
        ]
        return random.choice(jokes)

    def _get_weather(self):
        # Placeholder function, you can integrate with a weather API here
        return "I'm sorry, I can't provide real-time weather information."

    def _play_game(self):
        print("Let's play a guessing game! I'm thinking of a number between 1 and 100.")
        secret_number = random.randint(1, 100)
        attempts = 0
        while True:
            guess = int(input("Enter your guess: "))
            attempts += 1
            if guess < secret_number:
                print("Too low! Try again.")
            elif guess > secret_number:
                print("Too high! Try again.")
            else:
                print(f"Congratulations! You guessed the number in {attempts} attempts.")
                break

    def _ask_user_question(self, user_input=None):
        if user_input:
            return "What question would you like to ask?"
        else:
            question = input("Enter your question: ")
            return f"You asked: '{question}'. That's an interesting question!"

    def start_conversation(self):
        print(self.greet())
        while True:
            user_input = input("You: ")
            if user_input.lower() == 'exit':
                break
            response = self.handle_input(user_input)
            print("Chatbot:", response)

    def end_conversation(self):
        print(self.farewell)

if __name__ == "__main__":
    chatbot = AdvancedChatbot()
    chatbot.start_conversation()
    chatbot.end_conversation()

# Simple-Chat-Bot Commands :--------------------

#          "who are you"
#          "Who are you"
#          "how are you"
#          "How are you"
#          "what's your name"
#          "What's your name"
#          "what's the time"
#          "What's the time"
#          "what's your favorite color"
#          "What's your favorite color"
#          "tell me a joke"
#          "Tell me a joke"
#          "what's the weather like"
#          "What's the weather like"
#          "play game"
#          "Play game"
#          "ask user question"
#          "Ask user question"
#          "exit"

```
